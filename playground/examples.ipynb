```python
from decimal import Decimal

from rich import print as pprint

from opengd77.converters import channel_to_dict, codeplug_to_csvs, dicts_to_csv
from opengd77.models import Codeplug, DigitalChannel, TalkerAlias

channel = DigitalChannel(
    name="Test Channel",
    rx_frequency=Decimal("145.000"),
    tx_frequency=Decimal("145.000"),
    timeslot_1_talker_alias=TalkerAlias.APRS | TalkerAlias.TEXT,
    timeslot_2_talker_alias=TalkerAlias.TEXT | TalkerAlias.APRS,
)
pprint(channel)
channel_dict = channel_to_dict(channel, number=1)
pprint(channel_dict)
channels_csv = dicts_to_csv([channel_dict], channel_dict.keys())
pprint(channels_csv)
codeplug = Codeplug(
    channels=[channel],
)
codeplug_csvs = codeplug_to_csvs(codeplug)
pprint(codeplug_csvs)
```
